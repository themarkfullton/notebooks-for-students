```python
import tensorflow as tf
import numpy as np
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
# This is SUPER BASIC. We've got one layer with one neuron in it.
model = Sequential([Dense(units=1, input_shape=[2])])
model.compile(optimizer="sgd", loss="mean_squared_error")
model.summary()
```

```python
# X has 2 columns and 6 rows
# The first column correlates to y using the relationship y=2x-1; the second column is 'random' (starts at 2, increases by 1)
X = np.array([[-1.0,2.0],[0.0,3.0],[1.0,4.0],[2.0,5.0],[3.0,6.0],[4.0,7.0]], dtype=float)
y = np.array([-3.0,-1.0,1.0,3.0,5.0,7.0], dtype=float)
```

```python
model.fit(X, y, epochs=500)
```

```python
# We're giving the model a row with [10.0, 20.0]
# Remember, the first column corresponds using 2x -1; the seond column is trivial.
# Given this, we'd expect 10 to give us 19
model.predict([[10.0,20.0]])
```

```python
# This gives us our weights
# The first two are the weights it's using for your columns
# Our model has picked up on the fact that the first column is important since it's assigned it the weight of 2.145...
# It's also realized that the second column isn't too helpful, since it's giving it -0.14594....
# The last number given is the bias: -0.5621....
model.get_weights()
```
